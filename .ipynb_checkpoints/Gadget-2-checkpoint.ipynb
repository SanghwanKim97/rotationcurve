```python
#f, axes=plt.subplots(1,3)
#f.set_size_inches((10,4))
a=readsnap('/home/sanghwan/simul/Gadget-2.0.7/run1/Gadget2/galaxy/yesdarkmatter/snapshot_090', 'pos', 'disk')
b=readsnap('/home/sanghwan/simul/Gadget-2.0.7/run1/Gadget2/galaxy/yesdarkmatter/snapshot_090', 'vel', 'disk')
c=readsnap('/home/sanghwan/simul/Gadget-2.0.7/run1/Gadget2/galaxy/nodarkmatter/snapshot_090', 'pos', 'disk')
d=readsnap('/home/sanghwan/simul/Gadget-2.0.7/run1/Gadget2/galaxy/nodarkmatter/snapshot_090', 'vel', 'disk')

A=a[0:,0] #Slicing array into a data set about x position
B=b[0:,1] #Slicing array into a data set about y velocity
print(np.shape(A),np.shape(B))
MT=np.vstack((A,B)) #Attach the sets into one matrix : (100,000, ) + (100,000, ) => dimension (2,100,000) 
print(np.shape(MT))
MT=MT.T #Reshape the matrix so we can treat only position and velocity again -> dimension (100,000, 2)
 
X=sorted(MT, key=lambda x:x[0]) #Sorting the matrix in ascending order
X=np.array(X) #Make X be numpy array again since sorted function makes it only a data set, not an array

C=c[0:,0]#Same procedures about Dark matter
D=d[0:,1]
MTR=np.vstack((C,D))
MTR=MTR.T
Y=sorted(MTR, key=lambda x:x[0])
Y=np.array(Y)

plt.figure(figsize=(10,5))
print('\n')

for i in range(0,100000, 100):
    averagex=np.mean(X[i:i+100,0]+15) #Making average position to analysis
    averagey=np.mean(X[i:i+100,1])/2  #Making average velocity upon the position. The scaler added and divided is to make the graph consistuent
    plt.plot(averagex, averagey, marker='.', color='black')
    plt.xlabel('X (kpc)')
    plt.ylabel('Velocity (km/s)', rotation=360, labelpad=40)
    averageX=np.mean(Y[i:i+100,0]+10)/5
    averageY=np.mean(Y[i:i+100,1])/2
    plt.ylim(0,150)
    plt.xlim(0,20)
    plt.plot(averageX, averageY, marker='.', linestyle='', color='red')
plt.plot(averageX, averageY, label='No Dark Matter', color='red', marker='.')
plt.plot(averagex, averagey, label='With Dark Matter', color='black', marker='.')
plt.legend()
```

```python
from pygadgetreader import *
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import HTML
a=readsnap('/home/sanghwan/simul/Gadget-2.0.7/run1/Gadget2/galaxy/yesdarkmatter/snapshot_090', 'pos', 'disk')
b=readsnap('/home/sanghwan/simul/Gadget-2.0.7/run1/Gadget2/galaxy/yesdarkmatter/snapshot_090', 'vel', 'disk')
c=readsnap('/home/sanghwan/simul/Gadget-2.0.7/run1/Gadget2/galaxy/nodarkmatter/snapshot_090', 'pos', 'disk')
d=readsnap('/home/sanghwan/simul/Gadget-2.0.7/run1/Gadget2/galaxy/nodarkmatter/snapshot_090', 'vel', 'disk')

A=a[0:,0]
B=b[0:,1]

N=np.arange(1,10)
M=np.arange(1,10)
MT=np.vstack((A,B))

MT=MT.T

X=sorted(MT, key=lambda x:x[0])
X=np.array(X)

C=c[0:,0]
D=d[0:,1]
MTR=np.vstack((C,D))
MTR=MTR.T
Y=sorted(MTR, key=lambda x:x[0])
Y=np.array(Y)

plt.figure(figsize=(10,5))
averagex=np.zeros((100)) #Saving data sets in different variable so we can plot this out of the for loop 
averagey=np.zeros((100))
averageX=np.zeros((100))
averageY=np.zeros((100))
j=0
print('\n')
for i in range(0,100000, 1000):
    averagex[j]=np.mean(X[i:i+1000,0]+14.6)
    averagey[j]=np.mean(X[i:i+1000,1])/1.7
    #plt.plot(averagex, averagey, marker='.', color='black')
    plt.xlabel('X (kpc)')
    plt.ylabel('Velocity (km/s)', rotation=360, labelpad=40)
    averageX[j]=np.mean(Y[i:i+1000,0]+7)/5
    averageY[j]=np.mean(Y[i:i+1000,1])/2
    plt.ylim(0,150)
    plt.xlim(0,20)
    j=j+1
    #plt.plot(averageX, averageY, marker='.', linestyle='', color='red')
plt.plot(averageX, averageY, label='No Dark Matter', color='red', marker='.', linestyle='solid')
plt.plot(averagex, averagey, label='With Dark Matter', color='black', marker='.', linestyle='solid')
plt.legend()
plt.show()
```
